### Price

```python
restaurants.Price.unique()
```

```python
#pattern = re.compile('[¥€£₩฿₺₫]')
#restaurants_df['Price'] = restaurants_df.Price.apply(lambda s: pd.Series(pattern.sub('$', str(s))))
restaurants['Price'] = restaurants.Price.str.replace('[¥€£₩฿₺₫]', '$', regex=True)
```

```python
def set_category(price):
    if price == '$':
        return 'Muy Barato'
    elif price == '$$':
        return 'Accesible'
    elif price == '$$$':
        return 'Normal'
    elif price == '$$$$':
        return 'Costoso'
    elif price == '$$$$$':
        return 'Muy Costoso'
    else:
        return None

restaurants['PriceCategory'] = restaurants.Price.apply(set_category)
```
