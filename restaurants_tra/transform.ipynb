```python
import pandas as pd
import re
```

```python
# read data
original_df = pd.read_csv('../ods/michelin_my_maps.csv', dtype={'PhoneNumber': str})
original_df.head()
```

```python
# make a copy of the original data frame
restaurants_df = original_df.copy(deep=True)  # deep = True indicate 
```

```python
# raname a column
restaurants_df.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
```

```python
restaurants_df.columns
```

```python
restaurants_df.set_index('id', inplace=True)
```

```python
restaurants_df.shape
```

### Location and Country

```python
s = pd.Series([" A str series   "])
s
```

```python
s.str.strip().str.split()
```

```python
restaurants_df.Location.str.strip().str.split(",")
```

```python
restaurants_df[['Location2', 'Country']] = restaurants_df.Location.apply(lambda s: pd.Series(s.strip().split(",")))
restaurants_df.head()
```

### Price

```python
restaurants_df.Price.unique()
```

```python
pattern = re.compile('[¥€£₩฿₺₫]')
```

```python
restaurants_df['Price'] = restaurants_df.Price.apply(lambda s: pd.Series(pattern.sub('$', str(s))))
```

```python
restaurants_df.sample(n=5)
```

```python
def set_category(price):
    if price == '$':
        return 'Muy Barato'
    elif price == '$$':
        return 'Accesible'
    elif price == '$$$':
        return 'Normal'
    elif price == '$$$$':
        return 'Costoso'
    elif price == '$$$$$':
        return 'Muy Costoso'
    else:
        return None

restaurants_df['PriceCategory'] = restaurants_df.Price.apply(set_category)
```

```python
restaurants_df.sample(n=20)
```

### Cuisine

```python
restaurants_df[restaurants_df.Cuisine.str.contains('/')].shape
```

```python
# replacing / by ,
restaurants_df['Cuisine'] = restaurants_df.Cuisine.apply(lambda s: pd.Series(s.replace('/',',')))
```

```python
restaurants_df[restaurants_df.Cuisine.str.contains('/')].shape
```

```python
restaurants_df.Cuisine.apply(lambda s: pd.Series(len(s.strip().split(',')))).value_counts()
```

```python
restaurants_df[['Cuisine1', 'Cuisine2']] = restaurants_df.Cuisine.apply(lambda s: pd.Series(s.strip().split(',')))
```

```python
restaurants_df.head()
```
